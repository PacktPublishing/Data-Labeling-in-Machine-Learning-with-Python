```python
  # Calculate the RMS energy
rms = librosa.feature.rms(y=data)[0]

# Reference level for 0 dB (threshold of hearing)
reference_level = 20e-6

# Convert the RMS levels to dB using the standard reference level
loudness_in_db = 20 * np.log10(rms / reference_level)
print(loudness_in_db)
```

```python
#Visualize the loudness
 # Plot the loudness in dB across time
plt.figure(figsize=(10, 4))
plt.plot(librosa.times_like(rms, sr=sr), loudness_in_db)
plt.title('Loudness (dB) Across Time')
plt.xlabel('Time (s)')
plt.ylabel('Loudness (dB)')
plt.ylim(0, 120)  # Set the y-axis limits to represent the scale from hearing threshold to pain threshold
plt.show()
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load an audio file
audio_file = "cat_1.wav"
y, sr = librosa.load(audio_file)

# Calculate loudness using the RMS (Root Mean Square) energy
loudness = librosa.feature.rms(y=y)

# Convert the loudness to dB (decibels)
loudness_db = librosa.power_to_db(loudness)

# Create a loudness curve plot
plt.figure(figsize=(12, 4))
librosa.display.waveshow(loudness_db, sr=sr, x_axis='time')
plt.title("Loudness Curve")
plt.xlabel("Time (s)")
plt.ylabel("Loudness (dB)")
plt.show()
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load an audio file
audio_file = "cat_1.wav"
y, sr = librosa.load(audio_file)
# Extract the tempo 

tempo, _ = librosa.beat.beat_track(y=y, sr=sr) 

print(f"Tempo: {tempo} BPM") 

 
```

```python
 

# Calculate chroma feature 

chroma = librosa.feature.chroma_stft(y=y, sr=sr) 

  

# Display the chromagram 

plt.figure(figsize=(12, 4)) 

librosa.display.specshow(chroma, y_axis='chroma', x_axis='time') 

plt.title("Chromagram") 

plt.colorbar() 

plt.show() 
```

```python
# Calculate MFCC 

mfccs = librosa.feature.mfcc(y=y, sr=sr) 

  

# Display the MFCCs 

plt.figure(figsize=(12, 4)) 

librosa.display.specshow(mfccs, x_axis='time') 

plt.title("MFCCs") 

plt.colorbar() 

plt.show() 
```

```python
# Calculate zero-crossing rate 

zero_crossings_rate = librosa.feature.zero_crossing_rate(y) 

plt.figure(figsize=(12, 4)) 

plt.semilogy(zero_crossings_rate.T) 

plt.title("Zero-Crossing Rate") 

plt.show() 
```

```python
 

# Calculate spectral contrast 

spectral_contrast = librosa.feature.spectral_contrast(y=y, sr=sr) 

  

# Display the spectral contrast 

plt.figure(figsize=(12, 4)) 

librosa.display.specshow(spectral_contrast, x_axis='time') 

plt.title("Spectral Contrast") 

plt.colorbar() 

plt.show() 
```
