Using summary statistics of training dataset to generate housing price labels for regression.

1.predict the median_house_value(targte label) of unlabelled dataset by 
computing the summary statistics of training data for each target label. 
2. Then calculate the distance between unlabelled data set's each feature 
to the summary statistics(aggregated value) of corresponding feature in the training dataset.
3.finally assign the label to the unlabelled dataset with label from training dataset that has minimum distance from the summary statistics(aggregated value) of training dataset.


```python

import pandas as pd

# Load the labeled data
df = pd.read_csv('housing_training.csv')
```

```python
# Load the labeled data
df_test = pd.read_csv('housing.csv')
```

```python
# Compute the summary statistics for each feature by target label
summary_stats = df.groupby('median_house_value').describe()
```

```python
summary_stats
```

```python
import pandas as pd

# Load the labeled data
#df_labeled = pd.read_csv('labeled_data.csv')
df_labeled = df

# Compute the summary statistics for each feature by target label
#summary_stats = df_labeled.groupby('median_house_value').describe()

# Load the unlabeled data
#df_unlabeled = pd.read_csv('unlabeled_data.csv')
df_unlabeled = df_test

# Predict the target label for each data point in the unlabeled data
for i, row in df_unlabeled.iterrows():
    # Compute the distances to each target label's summary statistics
    dists = {}
    for median_house_value in summary_stats.index:
        print(median_house_value)
        dist = 0
        for col in df_unlabeled.columns:
            if col != 'median_house_value':
                dist += (row[col] - summary_stats.loc[median_house_value , (col, 'mean')]) ** 2
        dists[median_house_value] = dist
    print(dists)
    # Select the target label with the minimum distance
    predicted_target = min(dists, key=dists.get)
    print(predicted_target)
    # Assign the predicted target label to the current row
    df_unlabeled.at[i, 'median_house_value'] = predicted_target
```
